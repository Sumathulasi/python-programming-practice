## **Sales Data Analysis**
## **Project Idea**

**Analyze a sales dataset to find:**

*Total sales*

*Best-selling product*

*Monthly sales trend*

*Average sales*

**Tools:**

*Python*

*NumPy (numerical calculations)*

*Pandas (data analysis)*

*Matplotlib (visualization)*

```python
import pandas as pd
import numpy as np

# Sample data
data = {
    "Product": ["A", "B", "C", "A", "B", "C"],
    "Sales": [100, 150, 200, 120, 170, 210]
}

df = pd.DataFrame(data)

# Total sales
total_sales = np.sum(df["Sales"])
print("Total Sales:", total_sales)

# Average sales
average_sales = np.mean(df["Sales"])
print("Average Sales:", average_sales)

# Sales by product
sales_by_product = df.groupby("Product")["Sales"].sum()
print(sales_by_product)
```

## **count number of vowels in a given word**

```python
word= input('Hello, Provide a word , I will tell you how many vowels in it?')
count = 0
vowel =['a','e','i','o','u','A','E','I','O','U']
for x in word:
  if x in vowel:
    count += 1
print(count)
```

```python
def count_vowels(word):
    vowels = "aeiouAEIOU"
    count = 0
    for letter in word:
        if letter in vowels:
            count += 1
    return count

print(count_vowels("education"))
```

## **counting number of consonants in a given string**

```python
word= input('Hello, Provide a word , I will tell you how many consonants in it?')
count = 0
vowel =['a','e','i','o','u','A','E','I','O','U']
for x in word:
  if x not in vowel:
    count += 1
print(count)
```

```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(10)
```

## find middle element in a list

```python
my_list=[4,976,8,4]
length = len(my_list)
mid_elements = []

if length % 2 == 1:
  # Odd length, one middle element
  mid_elements = [my_list[length // 2]]
else:
  # Even length, two middle elements
  mid_elements = [my_list[length // 2 - 1], my_list[length // 2]]

print(mid_elements)
```

## compare two strings for anagram

```python
str1='sumadevi'
str2= 'umavides'
str1= str1.upper()
str2= str2.upper()
if (sorted(str1)== sorted(str2)):
  print('anagram')
else:
  print('not anagram')
```

```python
def anagram(str1,str2):
  str1= str1.upper()
  str2= str2.upper()
  if (sorted(str1)== sorted(str2)):
    print('anagram')
  else:
    print('not anagram')
```

## check for palindrome

```python
def palindrome(str):
  if str.lower() == str.lower()[::-1]:
    print('True')
  else:
    print('False')
```
